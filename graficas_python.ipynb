Grafica de una lista

```python
# Importamos la libreria
# import matplotlib.pyplot

from matplotlib.pyplot import plot, show, xlim, ylim, xticks, figure, subplot
```

```python
# Grafica No1

y = [1,2,3,4,5]

plot(y)
show()
```

```python
# Grafica N째 2
y = [1,2,3,4,5]

plot(y, '*')
show()
```

```python
# Grafica N째3
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '*')
show()
```

```python
# Grafica N째4

# Graficacon limites

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '*')
xlim(-10,10)
ylim(0, 15)
show()
```

```python
# Grafica N째5

from numpy import linspace, pi, sin
x = linspace(-10*pi, 10*pi, 2560)
y = sin(x)

plot(x,y)
ylim(-2,2)
show()
```

```python
# Grafica N째6

from numpy import linspace, pi, sin
x = linspace(-pi, pi, 256)
y = sin(x)

plot(x,y)
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

show()
```

```python
# Grafica N째7

from numpy import linspace, pi, sin, cos
from matplotlib.pyplot import plot, show, xlim, ylim, xticks, legend

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

plot(x,y, label = 'seno')
plot(x,z, label = 'coseno')
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

show()
```

```python
# Grafica N째8

from numpy import linspace, pi, sin, cos
from matplotlib.pyplot import plot, show, xlim, ylim, xticks, legend

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

# Figura 1
a = figure(2)
plot(x,y, label = 'seno')
legend(loc='upper left')
# Figura 20
b = figure(20)
plot(x,z, label = 'coseno')
legend(loc='upper left')
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

show()
```

```python
# Grafica N째9

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

# Subgraficas
subplot(2,1,1)
plot(x,y, label = 'seno')
legend(loc='upper left')
subplot(2,1,2)
plot(x,z, label = 'coseno')
legend(loc='upper left')
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])

show()
```

```python
# # Se importan las librer챠as necesarias

from numpy import linspace, sin, cos

from pylab import figure, show

from mpl_toolkits.mplot3d import Axes3D

# # generando algunos datos

x = linspace(-10,10,100);

y = sin(x);

z = cos(x);

# # Se crea la instancia de la figura y se asocia a la figura 3D

fig = figure()

ax = Axes3D(fig)

# # graficando los datos

for i in range(len(x)):

  ax.plot([x[i], x[i]], [y[i], y[i]], [0, z[i]], 

          '--', linewidth=2, color='b', alpha=.5)

# # graficando un circulo en el tope de cada curva

# plotting a circle on the top of each stem

ax.plot(x, y, z, 'o', markersize=8, 

        markerfacecolor='none', color='b',label='ib')

ax.set_xlabel('x')

ax.set_ylabel('y')

ax.set_zlabel('z')

show()
```
